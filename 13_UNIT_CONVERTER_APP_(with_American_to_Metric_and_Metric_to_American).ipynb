```python
#_________Length Conversions
def inches_to_cm(inches):
    return inches * 2.54

def cm_to_inches(cm):
    return cm / 2.54

def feet_to_meters(feet):
    return feet * 0.3048

def meters_to_feet(meters):
    return meters / 0.3048

def miles_to_kilometers(miles):
    return miles * 1.6093

def kilometers_to_miles(km):
    return km / 1.6093

#_________Weight Conversions
def pounds_to_kilograms(pounds):
    return pounds * 0.4535

def kilograms_to_pounds(kg):
    return kg / 0.4535

def ounces_to_grams(ounces):
    return ounces * 28.43

def grams_to_ounces(grams):
    return grams / 28.43

#_________Volume Conversions
def gallons_to_liters(gallons):
    return gallons * 3.78

def liters_to_gallons(liters):
    return liters / 3.78

#_________Temprature Conversions
def fahrenheit_to_celsius(f):
    return (f - 32) * 5/9

def celsius_to_fahrenheit(c):
    return (c * 9/5) + 32

#Now define the main control functions
def length_conversion():
    while True:
        print("\n_________Length Conversions")
        print("1. inches to centimeter")
        print("2. centimeter to inch")
        print("3. feet to meters")
        print("4. meters to feet")
        print("5. miles to kilometer")
        print("6. kilometer to miles")
        print("7. Back to main menu")

        choice = input("\nChoose an option (1-7): ")

        if choice == '1':
            inches = float(input("Enter inches: "))
            print(f"\n{inches} inches = {inches_to_cm(inches)} centimeters")
        elif choice == '2':
            cm = float(input("Enter centimeter: "))
            print(f"\n{cm} centimeter = {cm_to_inches(cm)} inches")

        elif choice == '3':
            feet = float(input("Enter feet: "))
            print(f"\n{feet} feet = {feet_to_meters(feet)} meters")

        elif choice == '4':
            meters = float(input("Enter meters: "))
            print(f"\n{meters} meters = {meters_to_feet(meters)} feet")

        elif choice == '5':
            miles = float(input("Enter miles: "))
            print(f"\n{miles} miles = {miles_to_kilometers(miles)} km")

        elif choice == '6':
            km = float(input("Enter kilometer: "))
            print(f"\n{km} kilometer = {kilometers_to_miles(km)} miles")

        elif choice == '7':
            break

        else:
            print("/nInvalid choice. Try again!")


def weight_conversion():
    while True:
        print("_________Weight Conversions")
        print("1. pounds to kilogram")
        print("2. kilogram to pounds")
        print("3. ounces to gram")
        print("4. gram to ounces")
        print("5. Back to Main Menu")

        choice = input("\nChoose an option (1-5): ")

        if choice == '1':
            pounds = float(input("Enter pound: "))
            print(f"\n{pounds} pounds = {pounds_to_kilograms(pounds)} kilogram")

        elif choice == '2':
            kg = float(input("Enter kilogram: "))
            print(f"\n{kg} kilogram = {kilograms_to_pounds(kg)} pounds")

        elif choice == '3':
            ounces = float(input("Enter ounces: "))
            print(f"\nf{ounces} ounces = {ounces_to_grams(ounces)} grams")

        elif choice == '4':
            grams = float(input("Enter grams: "))
            print(f"\n{grams} grams = {grams_to_ounces(grams)} ounces")

        elif choice == '5':
            break

        else:
            print("/nInvalid choice. Try again!")


def volume_conversion():
    while True:
        print("\n_________Volume Conversions")
        print("1. Gallons to Liters")
        print("2. Liters to Gallons")
        print("3. Back to Main Menu")

        choice = input("\nChoose an option (1-3): ")

        if choice == '1':
            gallons = float(input("Enter gallons: "))
            print(f"\n{gallons} gallons = {gallons_to_liters(gallons)} liters")

        elif choice == '2':
            liters = float(input("Enter liters: "))
            print(f"\n{liters} liters = {liters_to_gallons(liters)} gallons")

        elif choice == '3':
            break

        else:
            print("/nInvalid choice. Try again!")


def temperature_conversion():
    while True:
        print("_________Temprature Conversions")
        print("1. Fahrenheit to Celsius")
        print("2. Celsius to Fahrenheit")
        print("3. Back to Main Menu")

        choice = input("\nChoose an option (1-3): ")

        if choice == '1':
            f = float(input("Enter Fahrenheit: "))
            print(f"\n{f} fahrenheit = {fahrenheit_to_celsius(f)} celsius")

        if choice == '2':
            c = float(input("Enter Celsius: "))
            print(f"\n{c} celsius = {celsius_to_fahrenheit(c)} fahrenheit")

        if choice == '3':
            break

        else:
            print("/nInvalid choice. Try again!")


#Time to build the Main Menu
def main():
    while True:
        print("\n___Welcome to the UNIT CONVERTER APP___")
        print("1. Length Conversions")
        print("2. Weight Conversions")
        print("3. Volume Conversions")
        print("4. Temprature Conversions")
        print("5. Exit")

        choice = input("\nChoose an option (1-5): ")

        if choice == '1':
            length_conversion()

        elif choice == '2':
            weight_conversion()

        elif choice == '3':
            volume_conversion()

        elif choice == '4':
            temperature_conversion()

        elif choice == '5':
            print("\nThank you for using the Unit Converter. Goodbye, bro!")
            break

        else:
            print("\nInvalid choice. Try again!")

#Start the APP
main()

```
