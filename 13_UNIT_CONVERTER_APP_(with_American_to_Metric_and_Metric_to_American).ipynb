```python
# Imperial <-> Metric systems
# ===========================


# __________Length Conversions

def inches_to_cm(inches):
    return inches * 2.54    # 1inch = 2.54cm

def cm_to_inches(cm):
    return cm / 2.54

def feet_to_meters(feet):
    return feet * 0.3048     # 1feet = 0.3048m

def meters_to_feet(meters):
    return meters / 0.3048

def miles_to_kilometers(miles):
    return miles * 1.6093     # 1mile = 1.6093km

def kilometers_to_miles(km):
    return km / 1.6093

# __________Weight Conversions

def pounds_to_kilograms(pounds):
    return pounds * 0.4535    # 1pound = 0.4535kg

def kilograms_to_pound(kg):
    return kg / 0.4535

def ounces_to_grams(ounces):
    return ounces * 28.34      # 1ounce = 28.43g

def grams_to_ounces(grams):
    return grams / 28.34

# __________Volume Conversions

def gallons_to_liters(gallons):
    return gallons * 3.78   # 1gallon = 3.78L

def liters_to_gallons(liters):
    return liters / 3.78

# __________Temperature Conversions

def fahrenheit_to_celcius(f):
    return (f - 32) * 5/9   #Formula

def celcius_to_fahrenheit(c):
    return (c * 9/5) + 32

#Now define the main control functions (Menus and Navigation)
def length_conversion():
    while True:
        print("\n __________Length Conversions")
        print("1. Inches to Centimeter")
        print("2. Centimeter to Inches")
        print("3. Feet to Meters")
        print("4. Meters to Feet")
        print("5. Miles to kilometers")
        print("6. Kilometers to Miles")
        print("7. Back to main menu")

        choice = input("\nChoose an option (1 - 7): ")

        if choice == '1':
            inches = float(input("Enter inches: "))
            print(f"\n{inches} inches = {inches_to_cm(inches)} centimeters")

        elif choice == '2':
            cm = float(input("Enter Centimeters: "))
            print(f"\n{cm} centimeters = {cm_to_inches(cm)} inches")

        elif choice == '3':
            feet = float(input("Enter feet: "))
            print(f"\n{feet} feet = {feet_to_meters(feet)} meters")

        elif choice == '4':
            meters = float(input("Enter meters: "))
            print(f"\n{meters} meters = {meters_to_feet(meters)} feet")

        elif choice == '5':
            miles = float(input("Enter miles: "))
            print(f"\n{miles} miles = {miles_to_kilometers(miles)} kilometers")

        elif choice == '6':
            km = float(input("Enter kilometers: "))
            print(f"\n{km} kilometers = {kilometers_to_miles(km)} miles")
        elif choice == '7':
            break
        else:
            print("\nInvalid choice. Try again!")


def weight_conversion():
    while True:
        print("\n __________Weight Conversion")
        print("1. Pounds to Kilograms")
        print("2. Kilograms to Pounds")
        print("3. Ounces to Grams")
        print("4. Grams to Ounces")
        print("5. Back to Main Menu")

        choice = input("\nChoose an option (1-5): ")

        if choice == '1':
            pounds = float(input("Enter pounds: "))
            print(f"\n{pounds} pounds = {pounds_to_kilograms(pounds)} kilograms")

        elif choice == '2':
            kg = float(input("Enter kilograms: "))
            print(f"\n{kg} kilograms = {kilograms_to_pound(kg)} pounds")

        elif choice == '3':
            ounces = float(input("Enter ounces: "))
            print(f"\n{ounces} ounces = {ounces_to_grams(ounces)} grams")

        elif choice == '4':
            grams = float(input("Enter grams: "))
            print(f"\n{grams} grams = {grams_to_ounces(grams)} ounces")

        elif choice == '5':
            break

        else:
            print("\nInvalid choice. Try again!")


def volume_conversion():
    while True:
        print("\n __________Weight Conversion")
        print("1. Gallons to Liters")
        print("2. Liters to Gallons")
        print("3. Back to Main Menu")

        choice = input("\nChoose an option (1-3): ")

        if choice == '1':
            gallons = float(input("Enter gallons: "))
            print(f"\n{gallons} gallons = {gallons_to_liters(gallons)} liters")

        elif choice == '2':
            liters = float(input("Enter litres: "))
            print(f"\n{liters} litres = {liters_to_gallons(liters)} gallons")

        elif choice == '3':
            break

        else:
            print("\nInvalid choice. Try again!")



def temperature_conversion():
    while True:
        print("\n __________Tempratute Conversion")
        print("1. Fahrenheit to Celcius")
        print("2. Celcius to Fahrenheit")
        print("3. Back to Main Menu")

        choice = input("Choose an option (1-3): ")

        if choice == '1':
            f = float(input("Enter Fahrenheit: "))
            print(f"\n{f}Â°F = {fahrenheit_to_celcius(f)} Â°C")

        elif choice == '2':
            c = float(input("Enter celcius: "))
            print(f"\n{c}Â°C = {celcius_to_fahrenheit(c)} Â°F")

        elif choice == '3':
            break

        else:
            print("\nInvalid choice. Try again!")


#Time to build the Main Menu
def main():
    while True:
        print("\n___Welcome to the Unit Converter App___")
        print("1. Length Conversion")
        print("2. Weight Conversion")
        print("3. Volume Conversion")
        print("4. Temprature Conversion")
        print("5. Exit")

        choice = input("Choose an option (1-5): ")

        if choice == '1':
            length_conversion()

        elif choice == '2':
            weight_conversion()

        elif choice == '3':
            volume_conversion()

        elif choice == '4':
            temperature_conversion()

        elif choice == '5':
            print("\nThank you for using the Unit Converter. Goodbye, bro! ðŸ’ªðŸ”¥")
            break

        else:
            print("\nInvalid choice. Try again!")


#Start the APP
main()




```
