```python
from PIL import Image, ImageEnhance
from google.colab import files

# Upload an image
uploaded = files.upload()
image_path = list(uploaded.keys())[0]
img = Image.open(image_path)

# Functions to manipulate the image
def crop_image(img, left, upper, right, lower):
    return img.crop((left, upper, right, lower))

def rotate_image(img, angle):
    return img.rotate(angle)

def adjust_brightness(img, factor):
    enhancer = ImageEnhance.Brightness(img)
    return enhancer.enhance(factor)

def adjust_contrast(img, factor):
    enhancer = ImageEnhance.Contrast(img)
    return enhancer.enhance(factor)

# Main CLI Loop
while True:
    print("\nChoose an operation:")
    print("1. Crop")
    print("2. Rotate")
    print("3. Adjust Brightness")
    print("4. Adjust Contrast")
    print("5. Exit")

    choice = input("Enter your choice (1-5): ")

    if choice == '1':
        print("Cropping - Enter pixel values")
        left = int(input("Left: "))
        upper = int(input("Top: "))
        right = int(input("Right: "))
        lower = int(input("Bottom: "))
        img = crop_image(img, left, upper, right, lower)
        img.save("cropped.jpg")
        files.download("cropped.jpg")
        print("‚úîÔ∏è Cropped image downloaded!")

    elif choice == '2':
        angle = float(input("Enter rotation angle (degrees): "))
        img = rotate_image(img, angle)
        img.save("rotated.jpg")
        files.download("rotated.jpg")
        print("‚úîÔ∏è Rotated image downloaded!")

    elif choice == '3':
        factor = float(input("Enter brightness factor (1.0 = normal): "))
        img = adjust_brightness(img, factor)
        img.save("brightened.jpg")
        files.download("brightened.jpg")
        print("‚úîÔ∏è Brightness adjusted and downloaded!")

    elif choice == '4':
        factor = float(input("Enter contrast factor (1.0 = normal): "))
        img = adjust_contrast(img, factor)
        img.save("contrasted.jpg")
        files.download("contrasted.jpg")
        print("‚úîÔ∏è Contrast adjusted and downloaded!")

    elif choice == '5':
        print("üëã Exiting Photo Editor. All edits done!")
        break

    else:
        print("‚ùå Invalid choice. Try again.")
```
