```python
#This list will hold all our tasks.
#Each task is a dictionary with a "Title" and a "Completed" status.
tasks = []

#This function just prints out the main menu.
#It's what the user sees to know what actions are available.
def show_menu():
    print("\n __TO-DO LIST APP__")
    print("1. View tasks")
    print("2. Add task")
    print("3. Mark task as completed")
    print("4. Delete task")
    print("5. Exit")

#This functions shows current list of tasks.
def view_tasks():
    if not tasks:
        print("No tasks found.")
    else:
        #Loop list and print each task with its index.
        for index, task in enumerate(tasks):
            status = "✅" if task["completed"] else "❌"
            print(f"{index + 1}. {task['title']} [{status}]")

#This function adds a new task to the list.
def add_task():
    title = input("Enter task title: ")
    task = {"title": title, "completed": False}
    tasks.append(task)  #Add new task
    print("Task added")

#This function marks a task as completed (Changes its status)
def complete_task():
    view_tasks()

    try:
        #Ask user which to mark as complete.
        # Subtract 1 because Python lists start at 0, but humans count from 1.
        task_num = int(input("Enter task number to mark as complete")) - 1
        tasks[task_num]["completed"] = True
        print("Task marked as completed.")

    except (IndexError, ValueError):
        #Catches two common user mistakes:
        #IndexError -> Task number out of range.
        #ValueError -> input wasn't a number
        print("Invalid task number.")

#This Function deletes a task permanently.
def delete_task():
    view_tasks()    #Show the task again for reference

    try:
        task_num = int(input("Enter the task number to delete: ")) - 1
        tasks.pop(task_num) #Remove list
        print("Task deleted.")

    except (IndexError, ValueError):
        print("Invalid task number.")

#This is the main loop of the app.
#It keeps running until the user exits.
while True:
    show_menu()
    choice = input("Choose an option (1-5): ")

    #Use if-elif-else to trigger the right function
    if choice == "1":
        view_tasks()
    elif choice == "2":
        add_task()
    elif choice == "3":
        complete_task()
    elif choice == "4":
        delete_task()
    elif choice == "5":
        print("Goodbye!")
        break
    else:
        print("Invalid choice please try again")





```
